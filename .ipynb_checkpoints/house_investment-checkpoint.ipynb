## Grabbing data
Sources: [quandl](https://www.quandl.com/data/FMAC) and [wikipedia](https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States)

```python
import quandl
import pandas as pd
import pickle


api_key = open('quandlapikey.txt','r').read()

# pulling an example from FMAC database and all tables from Wikipedia
df = quandl.get('FMAC/HPI_AK', authtoken=api_key)




def get_states():
    states = pd.read_html('https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States')
    return states[0][1][2:]

def grab_initial_state_data():
    # empty dataframe to join stuff to
    main_df = pd.DataFrame()
    
    for abbreviation in get_states():
        query = 'FMAC/HPI_' + str(abbreviation)
        df = quandl.get(query, authtoken=api_key)
        df.columns = [str(abbreviation)]
        
        # calculate percent change
        df = (df - df[abbreviation][0]) / df[abbreviation][0] * 100.0

        if main_df.empty:
            main_df = df
        else:
            main_df = main_df.join(df)
            
        with open('states.pickle', 'wb') as pickle_out:
            pickle.dump(main_df, file=pickle_out)

def grab_mortgage_30y():
    df = quandl.get('FMAC/MORTG', trim_start='1975-01-01')
    df.columns = ['MORTG']
    
    df = (df - df['MORTG'][0]) / df['MORTG'][0] * 100.0
    
    df = df.resample('M').mean()
    
    return df
```

## Adding other economic indicators


```python
def sp500_data():
    df = pd.read_csv('GSPC.csv',index_col=['Date'], parse_dates=True)
    
    df["Adj Close"] = (df["Adj Close"]-df["Adj Close"][0]) / df["Adj Close"][0] * 100.0
    df = df.resample('M').mean()
    df.rename(columns={'Adj Close':'SP500'}, inplace=True)
    df = df['SP500']
    return df

# def us_unemployment():
#     df = quandl.get("ECPI/JOB_G", trim_start="1975-01-01", authtoken=api_key)
#     df["Unemployment Rate"] = (df["Unemployment Rate"]-df["Unemployment Rate"][0]) / df["Unemployment Rate"][0] * 100.0
#     df = df.resample('M').mean()
#     return df

def us_gdp():
    df = quandl.get("BCB/4385", trim_start="1975-01-01", authtoken=api_key)
    df["Value"] = (df["Value"]-df["Value"][0]) / df["Value"][0] * 100.0
    df = df.resample('M').mean()
    df.rename(columns={'Value':'USGDP'}, inplace=True)
    df = df['USGDP']
    return df
```

```python
SP500 = sp500_data()
US_GDP = us_gdp()
# US_unemployment = us_unemployment()
US_MORG = grab_mortgage_30y()
HPI_data = pd.read_pickle('states.pickle')
HPI_bench = get_benchmark_HPI()

HPI = HPI_data.join([HPI_bench, US_MORG, US_GDP, SP500])
HPI.dropna(inplace=True)

if not os.path.exists('HPI.pickle'):
    HPI.to_pickle('HPI.pickle')
```

## Classifier with scikit-learn
We need to define a label

```python
housing_data = pd.read_pickle('HPI.pickle')

housing_data = housing_data.pct_change()
housing_data.head()
```

Above, we can see there are `NaN`s and an `inf` as values in the rows and we must get rid of them

```python
import numpy as np

housing_data.replace([np.inf, -np.inf], np.nan, inplace=True)
housing_data.dropna(inplace=True)

housing_data.head(2)
```

```python
housing_data['USA_future'] = housing_data['USA'].shift(-1)
housing_data.dropna(inplace=True)

housing_data[['USA', 'USA_future']].head()
```

#### Define a label with a mapping function

```python
def create_labels(current, future):
    if future > current:
        return 1
    
    return 0

housing_data['label'] = list(map(create_labels, housing_data['USA'], housing_data['USA_future']))

housing_data.head()
```

```python
from sklearn import svm, preprocessing
from sklearn.model_selection import train_test_split

X = np.array(housing_data.drop(['USA_future', 'label'], axis=1))
X = preprocessing.scale(X)
y = np.array(housing_data['label'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, y_train)

print(classifier.score(X_test, y_test))
```

## How to discover an outlier

```python
import quandl
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

bridge_height = {'meters': [10.26, 10.31, 10.27, 10.22, 10.23, 6212.42, 10.28, 10.25, 10.31]}
df = pd.DataFrame(bridge_height)

df.plot(figsize=(20, 6))
plt.show()
```

### Comparison operators in pandas

```python
# define a rolling deviation and standard deviation
df['STD'] = df['meters'].rolling(2).std()
df_std = df.describe()['meters']['std']

df.plot(figsize=(20, 6))
plt.show()
```

```python
# leave datapoints only for data within x2 the standard deviation
df = df[(df['STD'] < df_std * 2)]

df['meters'].plot(figsize=(20, 6))
plt.show()
```
